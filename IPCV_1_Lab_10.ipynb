```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
import random
```

```python
img = cv2.imread('/content/rubiks.jpg',0)
cv2_imshow(img)
```

```python
row,col = img.shape
```

**Prewitts Operator**

```python
maskFx = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
maskFx
```

```python
maskFy = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
maskFy
```

```python
img_newFx = np.zeros([row, col])
```

```python
for i in range(1, row-1):
    for j in range(1, col-1):
        temp = img[i-1, j-1]*maskFx[0, 0]+img[i-1, j]*maskFx[0, 1]+img[i-1, j + 1]*maskFx[0, 2]+img[i, j-1]*maskFx[1, 0]+ img[i, j]*maskFx[1, 1]+img[i, j + 1]*maskFx[1, 2]+img[i + 1, j-1]*maskFx[2, 0]+img[i + 1, j]*maskFx[2, 1]+img[i + 1, j + 1]*maskFx[2, 2]
        img_newFx[i, j]= temp
```

*(1) Convolve original image with the Fx mask to get the x-gradient image.*

```python
cv2_imshow(img_newFx)
```

```python
img_newFy = np.zeros([row, col])
```

```python
for i in range(1, row-1):
    for j in range(1, col-1):
        temp = img[i-1, j-1]*maskFy[0, 0]+img[i-1, j]*maskFy[0, 1]+img[i-1, j + 1]*maskFy[0, 2]+img[i, j-1]*maskFy[1, 0]+ img[i, j]*maskFy[1, 1]+img[i, j + 1]*maskFy[1, 2]+img[i + 1, j-1]*maskFy[2, 0]+img[i + 1, j]*maskFy[2, 1]+img[i + 1, j + 1]*maskFy[2, 2]
        img_newFy[i, j]= temp
```

*(2) Convolve original image with the Fy mask to get the y-gradient image.*

```python
cv2_imshow(img_newFy)
```

*(3) Add the results of step (1) and (2).*

```python
Added = cv2.add(img_newFx,img_newFy)
cv2_imshow(Added)
```
