```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/deer2.jpg',0)
cv2_imshow(img)
```

```python
new = np.ravel(img)
```

```python
plt.hist(new)
```

```python
print(img.min(),img.max())
```

```python
minn = img.min()
maxx = img.max()
```

```python
row,col = img.shape
```

```python
k = (255-0)/(maxx-minn)
```

```python
for i in range(0,row-1):
   for j in range(0,col-1):
     pixel = img[i,j]
     pixel = k*(pixel-minn)

     img[i,j] = pixel
```

```python
cv2_imshow(img)
```
